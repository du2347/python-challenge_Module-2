```python
import csv

# Read the menu data and store it in a list
menu_file = "menu_data.csv"
menu = []

with open(menu_file, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    next(csvreader)  # Skip the header row

    for row in csvreader:
        menu.append(row)

# Read the sales data and create a report dictionary
sales_file = "sales_data.csv"
report = {}

with open(sales_file, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    next(csvreader)  # Skip the header row

    for row in csvreader:
        quantity = int(row[3])
        sales_item = row[4]

        if sales_item not in report:
            report[sales_item] = {
                "01-count": 0,
                "02-revenue": 0,
                "03-cogs": 0,
                "04-profit": 0,
            }

        for item in menu:
            if sales_item == item[0]:
                price = float(item[3])
                cost = float(item[4])
                profit = price - cost

                report[sales_item]["01-count"] += quantity
                report[sales_item]["02-revenue"] += price * quantity
                report[sales_item]["03-cogs"] += cost * quantity
                report[sales_item]["04-profit"] += profit * quantity
            else:
                print(f"{sales_item} does not equal {item[0]}! NO MATCH!")

# Print the report
for item, values in report.items():
    print(f"{item} {values}")

# Export the report to a text file
output_file = "pyramen_report.txt"
with open(output_file, "w") as textfile:
    textfile.write("PyRamen Sales Report\n")
    textfile.write("----------------------------\n")
    for item, values in report.items():
        textfile.write(f"{item} {values}\n")

print("Results have been exported to pyramen_report.txt")
```
